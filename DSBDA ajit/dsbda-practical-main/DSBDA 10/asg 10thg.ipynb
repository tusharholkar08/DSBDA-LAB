```python
import pandas as pd

# Load the Iris dataset
iris_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris = pd.read_csv(iris_url, names=iris_columns)

# Display the first few rows of the dataset
print(iris.head())

# Display feature names and data types
print(iris.dtypes)
```

```python
import matplotlib.pyplot as plt

# Plot histograms for each feature
iris.hist(figsize=(10, 8))
plt.suptitle('Histograms of Iris Dataset Features')
plt.show()
```

```python
# Plot box plots for each feature
iris.boxplot(figsize=(10, 6))
plt.title('Box Plots of Iris Dataset Features')
plt.show()
```

```python
# Identify outliers
outliers = []
for feature in iris.columns[:-1]:
    q1 = iris[feature].quantile(0.25)
    q3 = iris[feature].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers.extend(iris[(iris[feature] < lower_bound) | (iris[feature] > upper_bound)].index)

outliers = list(set(outliers))
print("Outliers:", outliers)
```
